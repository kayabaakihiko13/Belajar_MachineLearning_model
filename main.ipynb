<h3>Import Data</h3>

```python
%matplotlib widget
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.animation import FuncAnimation 
```

## Generate Data

```python
jumlah_data=100
y_data=np.array([i*0.1+np.random.randn() for i in range(jumlah_data)])
x_data=np.array([i*0.1 for i in range(jumlah_data)])
```

```python
plt.scatter(x_data,y_data)
plt.show()
```

# linear Gradien

```python
def gradien(x,m):
    return x*m
```

```python
x_prediction=np.array([0,10])
m_first=5
y_prediction=gradien(x_prediction,m_first)
```

# Modeling Model

```python
m_prediction=m_first
x_prediction_list=list()
y_prediction_list=list()
m_prediction_list=list()
learning_rate=0.1
for a in range(1,jumlah_data):
    y_prediction=np.array(gradien(x_data[a],m_prediction))
    y_actual=np.array(y_data[a])
    error=y_actual-y_prediction
    delta_gradien=error/x_data[a]

    m_prediction=m_prediction+delta_gradien
    m_prediction_list.append(m_prediction)
    y_prediction_list.append(gradien(np.array([0,10]),m_prediction))
    x_prediction_list.append(np.array([0,10]))
```

# Visualisasi Prediksi

```python
fig=plt.figure(figsize=(4,4))
line,=plt.plot([],[],c="#0A0101")
def animate(frame_num):
    x=x_prediction_list[frame_num]
    y=y_prediction_list[frame_num]
    line.set_data(x)
    line.set_data(y)
    return line,
plt.scatter(x_data,y_data)
plt.axis([0,10,0,10])
anime=FuncAnimation(fig,func=animate,frames=True,interval=100)
plt.show()
```
